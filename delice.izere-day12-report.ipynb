### Exercise 1:
Create a simple line plot for `y = x` for `x` values ranging from 0 to 10.


````python

x = np.linspace(0, 10, 10)
y = x 
plt.figure(figsize=(10, 10))
plt.plot(x, y, 'b')
plt.xlabel('x')
plt.ylabel('y')
plt.title(r"Plot of $y=x$")
plt.show()
````

Plot `y = x^2` using Matplotlib for `x` values between -10 and 10.

````python

x = np.linspace(-10, 10)
y = x ** 2
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'b')
plt.xlabel('x')
plt.ylabel('y')
plt.title(r"Plot of $y=x^2$")
plt.show()
````

### Exercise 3:
Create a scatter plot with 50 random points where x and y values are between 0 and 10.

````python

plt.figure(figsize=(8, 6), dpi=80)
n = 50
X = np.random.uniform(0,10,n)
Y = np.random.uniform(0,10,n)
plt.scatter(X,Y,color='y')

plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
````



### Exercise 4:
Plot a sine wave using Matplotlib for `x` values between 0 and 2.

````python

X = np.linspace(0, 2*np.pi, 50)
Y=  np.sin(X)
plt.figure(figsize=(8, 6))
plt.plot(X, Y)
plt.title(r"sine wave")
plt.show()
````

### Exercise 5:
Create a bar chart for the following data: `categories = ['A', 'B', 'C', 'D']` and `values = [3, 7, 5, 9]`.

````python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 9]
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color=['y'])
plt.title('Bar Chart of categories and values')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
````

### Exercise 6:
Create a histogram of 1000 random numbers generated from a normal distribution.

````python

mu, sigma = 0, 1 
data = np.random.normal(mu, sigma, 1000)
plt.hist(data, bins=30, color='red', alpha=0.7, edgecolor='black')
plt.title('Histogram of 1000 Random Numbers from Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
````

### Exercise 7:
Plot two line plots on the same graph: `y1 = x^2` and `y2 = x^3` for `x` ranging from 0 to 10.

````python
x_values = np.linspace(0, 10, 100)
y_squared = x_values**2
y_cubed = x_values**3
plt.plot(x_values, y_squared, label='y = x^2', color='blue')
plt.plot(x_values, y_cubed, label='y = x^3', color='orange')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid()
plt.show()
````

### Exercise 9:
Create a subplot with 2 rows and 1 column. Plot `y1 = x^2` in the first subplot and `y2 = x^3` in the second.

````python

input_values = np.linspace(-10, 10, 400)
output_square = input_values**2  # y = x^2
output_cube = input_values**3  # y = x^3
figure, axes = plt.subplots(2, 1, figsize=(8, 10))
axes[0].plot(input_values, output_square, color='blue')
axes[0].set_title(r'$y = x^2$', fontsize=16)
axes[0].set_xlabel('x', fontsize=14)
axes[0].set_ylabel('y', fontsize=14)
axes[0].grid()
axes[1].plot(input_values, output_cube, color='orange')
axes[1].set_title(r'$y = x^3$', fontsize=16)
axes[1].set_xlabel('x', fontsize=14)
axes[1].set_ylabel('y', fontsize=14)
axes[1].grid()
plt.tight_layout()
plt.show()
````

### Exercise 10:
Use a different color and linestyle for each line in a plot of three functions: `y1 = x`, `y2 = x^2`, and `y3 = x^3`.

````python

x = np.linspace(0, 10, 400)
y1 = x
y2 = x ** 2
y3 = x ** 3

plt.plot(x, y1, label='y = x', color='green', linestyle='-')
plt.plot(x, y2, label='y = x^2', color='blue', linestyle='--')
plt.plot(x, y3, label='y = x^3', color='red', linestyle='-.')
plt.title('Multiple Functions with Different Styles')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
````
